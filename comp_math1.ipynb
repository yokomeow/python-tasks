```python
from math import sin, pi
import matplotlib.pyplot as plt

L = 20
h = 0.5
NX = int(L / h + 1)
m = [i for i in range(1, NX + 1)]

T = 18
CFL = 1  #0.6, 1, 1.01
t = CFL * h
n = [i + 1 for i in range(int(T / t + 1))]

x = [(i - 1) * h for i in m]
y = [[sin(4 * pi / L * i) for i in x]]
yNx = y[0][len(y[0]) - 1]

for i in n:
    y.append([])
    y[i].append(yNx)
    for j in range(1, len(m)):
        y[i].append(y[i - 1][j] - CFL * (y[i - 1][j] - y[i - 1][j - 1]))
    yNx = y[i][len(y[i]) - 1]

u = [[sin(4 * pi / L * i) for i in x]]
uNx = u[0][len(u[0]) - 1]

for i in n:
    u.append([])
    u[i].append(uNx)
    for j in range(1, len(m) - 1):
        if j + 1 == len(m) - 1:
            u[i].append(u[i - 1][j] - CFL / 2 * (u[i - 1][1] - u[i - 1][j - 1]) + 
                    CFL ** 2 / 2 * (u[i - 1][1] - 2 * u[i - 1][j] + u[i - 1][j - 1]))
        u[i].append(u[i - 1][j] - CFL / 2 * (u[i - 1][j + 1] - u[i - 1][j - 1]) + 
                    CFL ** 2 / 2 * (u[i - 1][j + 1] - 2 * u[i - 1][j] + u[i - 1][j - 1]))
    uNx = u[i][len(u[i]) - 1]

'''

При CFL = 0,6  --> max i = 61; 
    CFL = 1    --> max i = 37; 
    CFL = 1.01 --> max i = 36. 
(Информация для проверки результатов на каждой итерации по времени)

'''
```

```python
i = 0
plt.plot(x, y[i])
plt.plot(x, u[i])
plt.show()
```
